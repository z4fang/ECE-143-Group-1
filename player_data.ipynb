```python
# importing library and read in data
import pandas as pd

nat_id = pd.read_csv('Data/player/nationality.csv');     # nationality id to nationality
player15 = pd.read_csv('Data/player/players_15.csv');
world_cup_matches = pd.read_csv('Data/world_cup_matches.csv');
world_cup_matches22 = pd.read_csv('Data/2022_world_cup_matches.csv');
```

#####                       <ins>Making List of country that was in world cup 2018 and 2022</ins>

```python
world_cup_matches.columns = [c.replace(' ', '_').lower() for c in world_cup_matches.columns]
world_cup_matches22.columns = [c.replace(' ', '_').lower() for c in world_cup_matches22.columns]
world_cup_matches22 = world_cup_matches22.drop(world_cup_matches22.index[48:64])  # drop rows without country name
world_cup_matches = world_cup_matches[world_cup_matches['year'] >= 2015]          # keeping only years bigger than 2015
world_cup_team = world_cup_matches.home_team.values.tolist()
world_cup_team = world_cup_team + world_cup_matches.away_team.values.tolist() + world_cup_matches22.away_team.values.tolist() # column to list
world_cup_team = [*set(world_cup_team)]             # remove duplicate
print(world_cup_team)

```

```python
# player15 = player15.drop(labels=['player_face_url', 'club_logo_url', 'club_flag_url', 'nation_logo_url', 'nation_flag_url',
#                                   'rb', 'gk','rcb','cb','lcb','lb','rwb','rdm','cdm','ldm','lwb', 'rm', 'rcm','cm','lcm',
#                                   'lm', 'ram', 'cam', 'lam', 'rw', 'rf', 'cf', 'lf', 'lw', 'rs', 'st', 'ls', ''], axis=1);

player15 = player15[['sofifa_id','short_name','short_name','long_name', 'value_eur', 'wage_eur', 'age', 'height_cm', 
                     'weight_kg','club_team_id', 'club_name','nationality_id', 'nationality_name']]

player15.nationality_name = player15.nationality_name.replace(['China PR', 'Republic of Ireland', 'Korea Republic'], ['China', 'Ireland', 'South Korea'])
player15 = player15[player15['nationality_name'].isin(world_cup_team)]
```
